```python
print(f'''Drugs: {drugs['Drug'].unique()}, 
Sex: {drugs['Sex'].unique()}
Blood Pressure: {drugs['BP'].unique()}
Cholesterol: {drugs['Cholesterol'].unique()}''')
```

```python
drugs = drugs.replace({'Sex':{'F':0, 'M':1}, 
                     'Cholesterol':{'NORMAL':0, 'HIGH':1},
                    'BP':{'LOW':0,'NORMAL':1, 'HIGH':2},
                    'Drug':{'drugA':0, 'drugB':1,'drugC':2,'drugX':3, 'DrugY':4 }})
drugs
```

```python
X = drugs.drop('Drug', axis=1)
y= drugs['Drug']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42, test_size = 0.3)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
model.fit(X_train, y_train)
```

```python
result = model.score(X_test, y_test)
print(f'Accuracy: {(result * 100):.2f}%')
```
